# Spirals
A playground for Archimidean and Fermat spirals. Inspired by shapes and patterns emerging from mathematical expressions, also found in nature, and probably not in Louis Vuitton and Herm√©s stores. 

## Archmidean spirals

```python
import numpy as np
import plotly.graph_objects as go

# Œ∏ range
theta = np.linspace(0, 8 * np.pi, 2000)

# Archimedean spiral
a1, b1 = 0, 0.2
r_arch = a1 + b1 * theta
x_arch = r_arch * np.cos(theta)
y_arch = r_arch * np.sin(theta)

# Fermat spiral
a2 = 0.8
r_ferm = a2 * np.sqrt(theta)
x_ferm = r_ferm * np.cos(theta)
y_ferm = r_ferm * np.sin(theta)

# Interactive plot
fig = go.Figure()

fig.add_trace(go.Scatter(
    x=x_arch, y=y_arch,
    mode='lines',
    name='Archimedean spiral'
))
fig.add_trace(go.Scatter(
    x=x_ferm, y=y_ferm,
    mode='lines',
    name='Fermat spiral'
))

fig.update_layout(
    title='Archimedean vs Fermat Spirals',
    xaxis=dict(showgrid=True, zeroline=True),
    yaxis=dict(showgrid=True, zeroline=True, scaleanchor='x', scaleratio=1),
    width=700, height=700,
    legend=dict(x=0.02, y=0.98)
)

fig.show()

```

```python
## Vogel spiral array
import numpy as np
import matplotlib.pyplot as plt

N = 1000
c = 5
alpha = np.deg2rad(137.507764)  # golden angle

n = np.arange(1, N+1)
r = c * np.sqrt(n)
theta = n * alpha

x = r * np.cos(theta)
y = r * np.sin(theta)

plt.figure(figsize=(6,6))
plt.scatter(x, y, s=10, color='royalblue')
plt.axis('equal')
plt.axis('off')
plt.title('Vogel (Phyllotactic) Spiral')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
N = 1000
c = 5
alpha = np.deg2rad(137.507764)  # golden angle in radians
n = np.arange(1, N + 1)

# Vogel spiral
r = c * np.sqrt(n)
theta = n * alpha
x = r * np.cos(theta)
y = r * np.sin(theta)

# Assign colors by modular grouping to show "arms"
num_arms = 21  # try 13, 21, 34 (Fibonacci numbers)
colors = np.mod(n, num_arms)

plt.figure(figsize=(7,7))
plt.scatter(x, y, c=colors, cmap='hsv', s=10)
plt.axis('equal')
plt.axis('off')
plt.title(f'Vogel Spiral with {num_arms} Colored Arms')
plt.show()
```

```python
import numpy as np
import plotly.graph_objects as go

# Parameters
N = 1000
c = 5
alpha = np.deg2rad(137.507764)  # golden angle
n = np.arange(1, N + 1)

# Vogel spiral points
r = c * np.sqrt(n)
theta = n * alpha
x = r * np.cos(theta)
y = r * np.sin(theta)

# Create base figure
fig = go.Figure()

# Main cloud
fig.add_trace(go.Scatter(
    x=x, y=y,
    mode='markers',
    marker=dict(size=4, color='lightgray'),
    name='Vogel points',
    visible=True
))

# Add several parastichy arms (Fibonacci gaps)
fib_gaps = [8, 13, 21, 34, 55]
for gap in fib_gaps:
    idx = np.arange(0, N - gap, gap)
    fig.add_trace(go.Scatter(
        x=x[idx],
        y=y[idx],
        mode='lines+markers',
        line=dict(width=2),
        name=f'Arm gap = {gap}',
        visible='legendonly'  # hidden by default
    ))

# Layout
fig.update_layout(
    title='Vogel Spiral with Parastichy Arms (Toggleable)',
    width=700, height=700,
    xaxis=dict(showgrid=True, zeroline=False, scaleanchor='y', scaleratio=1),
    yaxis=dict(showgrid=True, zeroline=False),
    legend=dict(x=0.02, y=0.98)
)

fig.show()
```
